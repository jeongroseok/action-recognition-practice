```python
import numpy as np
import cv2
import os
import matplotlib.pyplot as plt
```

```python
def visualize_img(img:np.ndarray):
    plt.axis('off')
    if len(img.shape) == 2:
        plt.imshow(img, cmap='gray')
    else:
        plt.imshow(img)
    plt.show()
```

```python
img = cv2.imread('./images/test.png')
visualize_img(img)
```

```python
img_bg = cv2.erode(img, np.ones((10, 10)), iterations=7)
img_l = img.copy()
img_l[:,:,1] -= img_bg[:,:,-1]
visualize_img(img_bg[:,:,1])
visualize_img(img-img_bg)
```
